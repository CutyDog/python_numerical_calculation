```python
def f(x,y):
    return x**2/5 + y**2/7

x = np.linspace(-5,5,300)
y = np.linspace(-5,5,300)
xmesh,ymesh = np.meshgrid(x,y)
z = f(xmesh,ymesh)

levels = [0.2,0.4,0.6,0.8,1]
colors = ["0.2","0.4","0.6","0.8"]
plt.contourf(x,y,z,colors=colors,levels=levels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y1 = np.sin(x)
y2 = np.sin(x-np.pi/2)

plt.plot(x,y1,color='r')
plt.plot(x,y2,color='b',linestyle='--')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y1 = np.sin(x)
y2 = np.sin(x-np.pi/2)

fig, axes = plt.subplots(2,1)
axes[0].set_ylim([-1.5,1.5])
axes[1].set_ylim([-1.5,1.5])
axes[0].plot(x,y1,color='r')
axes[1].plot(x,y2,color='g')

plt.show()
```

```python
def f(x1,y1):
    return x1**2 - y1**2/4

x1 = np.linspace(-5,5,300)
y1 = np.linspace(-5,5,300)
xmesh,ymesh = np.meshgrid(x1,y1)
z = f(xmesh,ymesh)


levels = [2,4,6,8,10]
colors = ["0.2","0.4","0.6","0.8"]
plt.contourf(x1,y1,z,colors=colors,levels=levels)

x = np.linspace(-5,5,300)

for i in x:
    plt.scatter(i,0,color='k')


plt.show()
```
