```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2+4*x-5

def df(x):
    return 2*x+8


def newton(f,df,a,eps,iter_max):
    iter = 0
    while True:
        a_new = a-f(a)/df(a)
        if abs(a-a_new)<eps:
            break
        a = a_new
        iter += 1

        if iter_max ==iter:
            break
    print("隗｣:",a_new,"險育ｮ怜屓謨ｰ:", iter)


```

```python
newton(f,df,6,0.001,1000)
```

```python
def f(x,y):
    return 5*x**2 + 6*x*y + 5*y**2 - 26*x - 26*y

def g(x,y):
    return x**2 + y**2 -4

x = np.linspace(-3,6,300)
y = np.linspace(-3,6,300)
xmesh,ymesh = np.meshgrid(x,y)
z1 = f(xmesh,ymesh)
z2 = g(xmesh,ymesh)



plt.contour(x,y,z2,colors='r',levels=[0])
plt.contour(x, y, z1, colors="k", levels=[-30,-15,0,15])
plt.show()
```
